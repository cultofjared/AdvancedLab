```python
%matplotlib inline
from __future__ import division
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as st
```

```python
n = 30
p = 0.5
std = np.sqrt(n * p * (1 - p))

# plot binomial
binomial_x = np.arange(0, n)
binomial_dist = st.binom(n, p)
binomial = binomial_dist.pmf(binomial_x)

binomial_fig, binomial_axis = plt.subplots()

# divide by n to compare to gaussian
binomial_x = binomial_x / n
binomial_x = binomial_x - 0.5
binomial_axis.plot(binomial_x, binomial);
```

```python
# recenter x points and plot gaussian
gauss_x = np.arange(-n/2, n/2)
gauss = st.norm(scale=std).pdf(gauss_x)

gauss_x = gauss_x / n
binomial_axis.plot(gauss_x, gauss);
binomial_fig
```

```python
diff = binomial - gauss
rel_diff = gauss / binomial
plt.ticklabel_format(axis='y', style='sci', scilimits=(-3,3))
plt.plot(diff**2);
```
